```python
import pandas as pd
```

```python
data = pd.read_csv('glue/sick/sick_original_zenodo.txt', delimiter='\t', header=0)
```

```python
data
```

```python
data = data[['pair_ID', 'relatedness_score', 'sentence_A', 'sentence_B']]
data.columns = ['old_index', 'score', 'sentence1', 'sentence2']
data['genre'] = 'main-captions'
data['filename'] = 'MSRvid'
data['year'] = '2012test'
data['source1'] = 'none'
data['source2'] = 'none'

data = data[['genre', 'filename', 'year', 'old_index', 'source1', 'source2', 'sentence1', 'sentence2', 'score']]
```

```python
data
```

```python
data_kaggle = pd.read_csv('glue/sick/sick_original_kaggle.txt', delimiter='\t', header=0)
train_ids = data_kaggle['pair_ID'].to_list()

data_train = data[data['old_index'].isin(train_ids)]
data_dev = data[~data['old_index'].isin(train_ids)]

# split data_dev into data_dev and data_test using 50/50 split
data_test = data_dev.sample(frac=0.5, random_state=42)
data_dev = data_dev.drop(data_test.index)

data_train.reset_index(drop=True, inplace=True)
data_dev.reset_index(drop=True, inplace=True)
data_test.reset_index(drop=True, inplace=True)
```

```python
# save to .tsv files
data_train.to_csv('glue/sick/train.tsv', sep='\t', index=True, header=True, index_label='index')
data_dev.to_csv('glue/sick/dev.tsv', sep='\t', index=True, header=True, index_label='index')
data_test.to_csv('glue/sick/test.tsv', sep='\t', index=True, header=True, index_label='index')
```
